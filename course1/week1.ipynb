```python
import numpy as np
def PatternCount(Text, Pattern):
    k = len(Pattern)
    count = 0
    for i in range(len(Text) - k + 1):
        if Text[i:i+k] == Pattern:
           count = count + 1 
    return count

def FrequentWords(Text, k):
    count = []
    for i in range(len(Text) - k + 1):
        count.append(PatternCount(Text, Text[i:i+k]))
    maxValue = max(count)
    maxValueIndexes = np.where(np.array(count) == maxValue)
    FrequentPatterns = set()
    for idx in maxValueIndexes[0]:
        FrequentPatterns.add(Text[idx: idx + k])
    return ' '.join(list(FrequentPatterns))
```

```python
import numpy as np

def FrequentWords(Text, k, t):
    FrequentPatterns = {}
    for i in range(len(Text) - k + 1):
        pattern = Text[i:i+k]
        if pattern not in FrequentPatterns:
            FrequentPatterns[pattern] = {}
            FrequentPatterns[pattern]['count'] = 0
            FrequentPatterns[pattern]['positions'] = []
        FrequentPatterns[pattern]['count'] = FrequentPatterns[pattern]['count'] + 1
        FrequentPatterns[pattern]['positions'].append(i)
    filtPatternsDict = {} 
    patterns = []
    for k, v in FrequentPatterns.items():
        if v['count']>=t:
            filtPatternsDict[k] = v
            patterns.append(k)
    return patterns

def ClumpFinding(genome, k, L, t):
    patterns = set()
    for i in range(len(genome) - L + 1):
        words = FrequentWords(genome[i:i+L], k, t)
        for word in words:
            patterns.add(word)
    #return ' '.join(list(patterns))
    return list(patterns)
```

```python
import numpy as np
def PatternToNumber(pattern):
    vocabDict = {'A': 0, 'C':1, 'G':2, 'T': 3}
    number = 0
    for i, c in enumerate(pattern[::-1]):
        number = number + vocabDict[c]*np.power(4, i)
    return number

def ComputingFrequencies(Text, k):
    FrequencyArray = np.zeros(np.power(4, k), dtype=int)
    for i in range(len(Text) - k +1):
        j = PatternToNumber(Text[i: i + k])
        FrequencyArray[j] = FrequencyArray[j] + 1
    FrequencyArray = np.array(FrequencyArray, dtype=str)
    return ' '.join(FrequencyArray)
```
